# DATA FRAMES

A DATA FRAME is a 2 dimensional data structure in which we store the data in form of "TABLES" i.e in the view of [rows X columns] ex: (4,5) # 4-->row indication, 5-->column indication

FEATURES OF DATA FRAMES :

-->potentially columns of different data types. -->Size is Mutable -->Its Contains only Labelled Data Axeses(Row,Column) -->we can perform Arithmetic Operations

"we can Create a Table via DATA FRAME that is known as DATA SET" WHAT IS DATA SET ? A DataSet is another name of table,in which we store the data in the form of Rows & Columns.The data is Mutable,Updatable.The data set Size is Negotiable.

STRUCTURE OF DATAFRAME/DATASET :

1.ROWS      
2.COLUMNS    
3.CELLS     

# SYNTAX OF DATA FRAME:     
pandas.DataFrame(data,index,columns,dtype,copy) DATA-->Info , INDEX-->Val ,COLUMNS--> Operations on it indicating , dtype--> Data type specifing , copy-->dublication

PANDAS COMPUTATIONAL TOOLS FOR STATISTICAL COMPUTATION OR MATHEMATICAL OPERATIONS --A computation is a process or completing a task by using many operations mathematically or statistically.

we have top 5 computational tools for Statistics

1.min() ---->which returns minimum value in a column.

2.max() ---->which returns maximum value in a column.

3.rank () ----> By using this method we can rank all the data Members in a column in a Ascending order if those are in Numerics,Come to Alphabets.It follows Alphabetic Order.

4.corr() --co-relation

    ----->It is a Relation between two Column Data Members in a Dataset.
         2.We use a Method called "corr()"
         3.It is a Scaled Form of a Covariance.
         4.The Correlation Values are lies between " -1 and +1 ".
         ----------
          SYNTAX: attribute1.corr(attribute2)

        TYPES OF CORRELATIONS:
a).POSITIVE(0 TO 1 in between values we got means positive)

    ---if values in the range of 0 to +0.5 then it said to be "High positive".  

   ---if values in the range of +0.5 to +1 then it is said to be "Low positive".
b).NEGATIVE(-1 to 0 in between we got means negative)

   ---if values in the range of 0 to -0.5 then it is said to be "High negative".  

   ---if values in the range of -0.5 to -1 then it is said to be "Low negative".   
3).NO CORRELATION

5.cov() --co-varience

       ---> THE RELATION BETWEEN TWO DATA MEMBERS WHICH ARE IN TWO DIFFERENT ATTRIBUTES.
---> IT IS A MEASUREMENT OF CORRELATION.

---> INDIVIDUAL ELEMENTS OF THE DATA FROM 2 COLUMNS.

---> THE RESULT MAY VARY EVERY SECOND.

# ------------------------------Different types of linestyle visuals-----------------------------



so here we have some terminologies to use in graph visualization they are:     
1. marker  --> in this we have only 'o' and '*' to indicate in graph             

2.linestyle ---> in this we have 
                  (i)  'dotted'   
                  (ii) 'dashed'    
                  (iii) 'solid'    
                  (iv) 'dashdot'     

# Line Plot with Different Linestyles

```python
import pandas as pd
import matplotlib.pyplot as plt
# Plotting line plots with different linestyles
plt.figure(figsize=(10, 6))
# Line plot with different linestyles
plt.plot(df['Exam_1_Score'], linestyle='-', label='Solid Line')
plt.plot(df['Exam_2_Score'], linestyle='--', label='Dashed Line')
plt.plot(df['Exam_3_Score'], linestyle='-.', label='Dash Dot Line')
plt.plot(df['Final_Exam_Score'], linestyle=':', label='Dotted Line')
plt.title('Different Linestyles - Line Plot')
plt.xlabel('Students')
plt.ylabel('Scores')
plt.legend()
plt.grid()
plt.show()
```

# Bar Plot with Different Linestyles (Edge Properties)

```python
import pandas as pd
import matplotlib.pyplot as plt
# Calculating the total score for each student
df['Total_Exams_Score'] = df['Exam_1_Score'] + df['Exam_2_Score'] + df['Exam_3_Score']

# Plotting the bar chart
plt.figure(figsize=(10, 6))
# Plotting the bar chart with customized edge properties (dashed)
bars = plt.bar(df.index, df['Total_Exams_Score'], edgecolor='black', linestyle='dashed')
plt.title('Aggregate Scores Comparison')
plt.xlabel('Students')
plt.ylabel('Total Scores')
# Custom legend for linestyle (since linestyle for bars isn't a direct parameter)
plt.legend([bars[0]], ['Total Scores'], loc='upper right')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
# Bar plot with different linestyles (using edge properties)
bars = plt.bar(df.index, df['Total_Exams_Score'], edgecolor='black', linewidth=1, linestyle='dotted')
plt.title('Different Linestyles - Bar Plot')
plt.xlabel('Students')
plt.ylabel('Total Scores')
plt.legend([bars[0]], ['Total Scores'], loc='upper right')
plt.show()
```

# Scatter Plot with Different Linestyles (Marker Edge)

```python
plt.figure(figsize=(10, 6))
# Scatter plot with different linestyles (using marker edge properties)
plt.scatter(df['Exam_1_Score'], df['Exam_2_Score'], edgecolor='black', linewidth=1, linestyle='dashdot', label='Scatter Plot')
plt.title('Different Linestyles - Scatter Plot')
plt.xlabel('Exam 1 Scores')
plt.ylabel('Exam 2 Scores')
plt.legend()
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.random.normal(10,20,10)    
print(x)
plt.hist(x,100)
plt.show()
```

# 2. UNIFORM DISTRIBUTION

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.random.uniform(10,20,10)    
print(x)
plt.hist(x,100)
plt.show()
```
